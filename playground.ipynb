# Initial inference (raw base model)

```python
from vllm import SamplingParams, LLM

llm = LLM(model='Qwen/Qwen2.5-1.5B-Instruct')
sampling_params = SamplingParams(temperature=0.8, top_p=0.95, max_tokens=512)
```

```python
prompt = "How many r's are in the word strawberry?"
SYSTEM_PROMPT = """
Respond in the following format:
<reasoning>
Your short work toward the answer.
</reasoning>
<answer>
Only the final number.
</answer>
"""
conversation = [
    {"role": "system", "content": SYSTEM_PROMPT},
    {"role": "user", "content": prompt},
]
outputs = llm.chat(conversation, sampling_params, use_tqdm=False)
```

```python
print(outputs[0].outputs[0].text)
```

# Finetuned

```python
from vllm import LLM, SamplingParams


llm_ft = LLM(model="KickItLikeShika/Qwen2.5-1.5B-Instruct-SFT-GRPO-GSM8K")
sampling_params_ft = SamplingParams(temperature=0.8, top_p=0.95, max_tokens=512)
```

```python
prompt = "How many r's are in the word strawberry?"
SYSTEM_PROMPT = """
Respond in the following format:
<reasoning>
Your short work toward the answer.
</reasoning>
<answer>
Only the final number.
</answer>
"""
conversation = [
    {"role": "system", "content": SYSTEM_PROMPT},
    {"role": "user", "content": prompt},
]
outputs_ft = llm_ft.chat(conversation, sampling_params_ft, use_tqdm=False)
```

```python
print(outputs_ft[0].outputs[0].text)
```
